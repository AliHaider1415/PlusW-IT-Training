```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Task 01 - Stock Market Trend Analysis (Line Plot)

```python
days = np.arange(1, 31)
prices = np.cumsum(np.random.randn(30) * 2 + 100)

plt.figure(figsize=(10, 5), dpi=100)
plt.plot(days, prices, linestyle='-', marker='o', color='b', markersize=6, label="Stock Price")
plt.xlabel("Day")
plt.ylabel("Price (USD)")
plt.title("Stock Price Trend")
plt.legend()
plt.grid(True)

plt.show()
```

### Task 02 - Customizing Temperature Trends with Line Styles and Markers

```python
# Generate days (1 to 10)
days = np.arange(1, 11)

# Generate random temperatures
temp_day = np.random.randint(25, 35, size=10)  # Day temperatures (25°C - 34°C)
temp_night = np.random.randint(15, 25, size=10)  # Night temperatures (15°C - 24°C)

# Create figure
plt.figure(figsize=(8, 5))

# Plot day temperature
plt.plot(days, temp_day, linestyle='--', marker='o', color='r', markersize=8, label="Day Temperature")

# Plot night temperature
plt.plot(days, temp_night, linestyle='-', marker='s', color='b', markersize=8, label="Night Temperature")

# Labels, title, legend, and grid
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Variation")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

### Task 03 - Population Growth Visualization (Scatter Plot)

```python
# Sample data
countries = ["USA", "China", "India", "Brazil", "Germany"]
pop_growth = np.array([0.8, 1.2, 1.5, 0.9, 0.5])  # Growth in percentage
population = np.array([331, 1441, 1380, 213, 83])  # Population in millions

# Create scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(pop_growth, population, c=population, cmap='viridis', s=population * 2, alpha=0.6)

# Labels and title
plt.xlabel("Population Growth (%)")
plt.ylabel("Total Population (millions)")
plt.title("Population Growth vs. Total Population")

# Add country annotations
for i, country in enumerate(countries):
    plt.annotate(country, (pop_growth[i], population[i]), fontsize=10, ha='right')

# Add colorbar and grid
plt.colorbar(label="Population (millions)")
plt.grid()

# Show the plot
plt.show()
```

### Task 04 - Customer Segmentation with Scatter Plot (Color Categories)

```python
# Simulated customer data
np.random.seed(42)  # Ensure reproducibility
age = np.random.randint(18, 60, 100)  # Generate 100 random ages (18-59)
spending_score = np.random.randint(20, 100, 100)  # Random spending score (20-99)
categories = np.random.choice([1, 2, 3, 4], 100)  # Assign 1 of 4 customer categories

# Create the scatter plot
plt.figure(figsize=(8, 5))
scatter = plt.scatter(age, spending_score, c=categories, cmap='jet', alpha=0.7)

# Labels and title
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation Analysis")

# Add color legend
plt.colorbar(label="Customer Category")

# Add grid and show plot
plt.grid()
plt.show()
```

### Task 05 - 3D Scatter Plot of Sales Data

```python
# Simulated sales data
np.random.seed(42)
region = np.random.randint(1, 6, 50)  # 5 regions (1 to 5)
sales = np.random.randint(500, 10000, 50)  # Sales between 500 and 10,000
profit = np.random.randint(50, 1000, 50)  # Profit between 50 and 1,000

# Create a 3D figure
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Create the scatter plot
scatter = ax.scatter(region, sales, profit, c=profit, cmap='coolwarm', s=50, alpha=0.7)

# Labels and title
ax.set_xlabel("Region")
ax.set_ylabel("Sales")
ax.set_zlabel("Profit")
ax.set_title("3D Sales Performance Analysis")

# Add colorbar
fig.colorbar(scatter, label="Profit")

# Show the plot
plt.show()
```
